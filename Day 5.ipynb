```python
import pandas as pd
```

```python
df_edata=pd.read_excel('edata.xlsx',sheet_name='Emp')
df_edata.head()
```

```python
#Data Exploration
#Handle the missing data in 'Age' and 'Dept' columns
#Age column

filter_age=df_edata['Age'].isnull()
df_edata[filter_age].head()



```

```python
filter_dept=df_edata['Dept'].isnull()
df_edata[filter_dept].head()
```

```python
#Handle missing data in 'Age' coumn with scikit learn
#Simple Imputer
import numpy as np
from sklearn.impute import SimpleImputer
imputer_age=SimpleImputer(missing_values=np.nan,strategy='mean')

imputer_age.fit(df_edata[['Age']])           #Search
```

```python
imputer_age.transform(df_edata[['Age']])    #Make changes
```

```python
imputer_age.fit_transform(df_edata[['Age']])   #Search and make changes
```

```python
transformed_age=imputer_age.fit_transform(df_edata[['Age']])
df_edata['Age']=transformed_age
df_edata.head()
```

```python
df_edata.info()
```

```python
#df_edata['Data']=df_edata['Data'].astype('Category')
df_edata['Gender']=df_edata['Gender'].astype('category')
df_edata['Seniority']=df_edata['Seniority'].astype('category')
```

```python
filled_dept=df_edata.fillna({'Dept':'unknown'})
df_edata['Dept']=filled_dept['Dept']
```

```python
df_edata.head()
```

```python
#Scaling using Standard Scaler
from sklearn.preprocessing import StandardScaler
sc_age = StandardScaler()
scaled_age = sc_age.fit_transform(df_edata[['Age']])
sc_salary = StandardScaler()
scaled_salary = sc_salary.fit_transform(df_edata[['Salary']])

df_edata['scaled_age'] = scaled_age
df_edata['scaled_salary'] = scaled_salary
```

```python
df_edata.head()
```

```python
#Scaling using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler_bonus = MinMaxScaler()
transformed_bonus = scaler_bonus.fit_transform(df_edata[['Bonus']])
#scaler_bonus.data_max_
#scaler_bonus.data_min_
#scaler_bonus.data_range_

df_edata['Bonus'] = transformed_bonus
df_edata.head()
```
