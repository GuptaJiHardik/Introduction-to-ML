```python
#seaborn is a library for making statistical graphics in Python. It is built on 
#top of matplotlib and closely integrated with pandas data structures.
#pip install seaborn
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
#Univariate Analysis
#Distribution plot
sns.distplot(tips['total_bill'],kde=False,bins=10)
```

```python
#Jointplot
#kind='hex' is used to plot hexagon
#kind='reg' is used to plot regression line
#kind='kde' is used to plot kernel density estimation
#kind='scatter' is used to plot scatter plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.pairplot(tips,hue='sex',palette='viridis')
```

```python
#categorical plot
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker',palette='coolwarm')
```

```python
#matrix plot
corr = tips.corr()
corr
```

```python
sns.heatmap(corr,cmap='PuRd',annot=True)    
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
matrix_p=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(matrix_p,cmap='coolwarm',linecolor='white',linewidths=1)
```

```python
sns.clustermap(matrix_p,cmap='coolwarm',standard_scale=1,annot=True)
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
import matplotlib.pyplot as plt
#grid plot
pair=sns.PairGrid(iris)
pair.map_diag(sns.histplot)
pair.map_upper(plt.scatter)
pair.map_lower(sns.kdeplot)
```
