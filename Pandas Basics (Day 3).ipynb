```python
df_emp=pd.read_csv("Day 3.csv")
df_emp.head()
```

```python
df_emp.info()
```

```python
df_emp['Start Date'] = pd.to_datetime(df_emp['Start Date'])
df_emp.info()
```

```python
df_emp['Last Login Time']=pd.to_datetime(df_emp['Last Login Time'])
df_emp.info()
```

```python
df_emp['Senior Management'] = df_emp['Senior Management'].astype('bool')
df_emp.info()
```

```python
df_emp['Gender'] = df_emp['Gender'].astype('category')
df_emp.info()
```

```python
#Pandas Filtering
#1

filter_male = df_emp['Gender'] == 'Male'
df_emp[filter_male]
```

```python
#2  To check the Finance member in the Team
filter_finance = df_emp['Team'] == 'Finance'
df_emp[filter_finance]
df_emp[filter_finance][['First Name','Start Date','Salary','Team']].head()
```

```python
#3  To check True in Senior Management
filter_sm = df_emp['Senior Management']
df_emp[filter_sm][['First Name','Start Date','Salary','Senior Management']].head()
```

```python
#4  To check False in Senior Management
filter_smf = df_emp['Senior Management'] !=  True
df_emp[filter_smf][['First Name','Start Date','Salary','Senior Management']].head()
```

```python
#5 List all employees whose bonus is less than 1.5
filter_bonus = df_emp['Bonus %'] < 1.5
df_emp[filter_bonus][['First Name','Start Date','Salary','Bonus %']].head()
#  df_emp[df_emp['Bonus %'] < 1.5][['First Name','Start Date','Salary','Bonus %']].head()
```

```python
# Salary more than 90000
df_emp[df_emp['Salary'] > 90000][['First Name','Start Date','Salary','Salary']].head()
```
