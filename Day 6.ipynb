```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1, 5, 11)
y = x ** 2

plt.plot(x, y, 'ro', label='x^2')         # Plot red circles for x^2
plt.plot(x, 2 ** y, 'bs', label='2^x')    # Plot blue squares for 2^x
```

```python
#Functional Method
plt.plot(x, y, color='red', marker='o', linestyle='--', linewidth=2)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Relationship between x and y')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.title('First Plot')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.title('Second Plot')

plt.show()
```

```python
#Object Oriented Method

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'b')

axes.set_xlabel('x label')
axes.set_ylabel('y label')

axes.set_title('Relationship between x and y')

plt.show()
```
